#### Show a visual summary of the data using boxplot. What Are the First and Third Quartiles? Comment your results.

* Hint : Quantiles vs Quartiles*
In Probability and Statistics, quantiles are cut points dividing the range of a probability distribution into continuous intervals with equal probabilities. When division is into four parts the values of the variate corresponding to 25%, 50% and 75% of the total distribution are called quartiles.

```python
# YOUR CODE HERE
# Visual summary using boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Title', y='Salary', data=employee)
plt.title('Boxplot of Salary by Title')
plt.xlabel('Title')
plt.ylabel('Salary')
plt.show()

# First and third quartiles
Q1 = employee['Salary'].quantile(0.25)
Q3 = employee['Salary'].quantile(0.75)
Q1, Q3
```

```python
# print first quartile here
Q1 = employee['Salary'].quantile(0.25)
Q1
''' mean = 48.88 since the mean of my salaries is almost 49, the first quartile is 35 that means 25% of the salaries are below my average salary.
and the third quartile is 60 that means 75% of the salaries are below 60. I have an outlier in my data which is 70 the highest salary in the company 
which they are VP's.'''
```

```python
# print third quartile here
Q3 = employee['Salary'].quantile(0.75)
Q3
```

#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
# your answer here
numeric_cols = employee.select_dtypes(include=[np.number]).columns
departments_stats = employee.groupby("Department")[numeric_cols].agg(['min', 'mean', 'max'])
departments_stats
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# your answer here
min_max_salary = employee.groupby("Department")["Salary"].agg(lambda x: x.max() - x.min())
min_max_salary
```
